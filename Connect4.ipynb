A brief documentation:

**Initialization:**
> the __init__(self) function takes no parameters
>> ex: board = Board()

**Resetting the board:**
> The resetBoard(self) function takes no parameters

> Clears the board object for reuse

> Returns an empty board
>> ex: board.resetBoard()

**Making moves:**
> The addToBoard(self, column, player) function takes two parameters: column and player.
>> Column is a number, [0, 6], that is a column that has not yet been filled.

>> Player is any character/string not equal to zero. Keep player chars consistent.

> Returns the updated board
>> ex: board.addToBoard(4, P1)

**Getting Possible Moves:**
> Takes no parameters

> Returns a list of valid columns
>> ex: moves = board.getPossibleMoves()


**Is game**
> isGame() returns True if no one has one and the board is not full
>> ex: game = board.isGame()

> isDraw() returns True if the game ended in a draw. Throws an error if board.isGame()
>> ex: draw = board.isDraw()

**Printing**
> display the board with display()
>> ex: board.display()
